```python

import pandas as pd

matches = pd.read_csv("PremierLeague.csv", index_col=0)

print(matches)

```

```python
matches.dtypes
```

```python
# matches["MatchWeek"].value_counts()
# matches["HomeTeam"].value_counts()
matches[matches["HomeTeam"] == "Liverpool"]
```

```python
matches["Date"] = pd.to_datetime(matches["Date"])
```

```python
# Hometeam
matches["HomeTeam_code"] = matches["HomeTeam"].astype("category").cat.codes 

# Awayteam
matches["AwayTeam_code"] = matches["AwayTeam"].astype("category").cat.codes

# Time (if time is not NAN)
# Check if 'Time' is not NaN before applying the transformation
matches.loc[matches["Time"].notna(), "Time"] = (
    matches.loc[matches["Time"].notna(), "Time"]
    .str.replace(":.+", "", regex=True)
    .astype(int)
)


```

```python
matches["date_code"] = matches["Date"].dt.dayofweek
```

```python
matches["target"] = (matches["FullTimeResult"] == "H").astype(int)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = matches[matches["Date"] < "2020-01-01"]
test = matches[matches["Date"] >= "2020-01-01"]

predictors = ["HomeTeam_code", "AwayTeam_code", "Time", "date_code"]

rf.fit(train[predictors], train["target"])
```

```python
predictions = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["target"], predictions)
```

```python
combined = pd.DataFrame(dict(actual=test["target"], prediction=predictions))

pd.crosstab(index=combined["actual"], columns=combined["prediction"])
```

```python
grouped_matches = matches.groupby("HomeTeam")

group = grouped_matches.get_group("Man City")

group
```

```python
def rolling_avg(group, cols, new_cols):
    group = group.sort_values("Date")
    rolling_stats = group[cols].rolling(3, closed="left").mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group


cols = [
    "FullTimeHomeTeamGoals", 
    "FullTimeAwayTeamGoals", 
    "HomeTeamShots",
    "HomeTeamShotsOnTarget",
    "HomeTeamFouls",
    "HomeTeamYellowCards",
    "HomeTeamRedCards",
    "B365HomeTeam",
]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_avg(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("HomeTeam").apply(lambda x: rolling_avg(x, cols, new_cols))

```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel("HomeTeam")
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
def make_predictions(data, predictors):
    train = data[data["Date"] < "2020-01-01"]
    test = data[data["Date"] >= "2020-01-01"]
    rf.fit(train[predictors], train["target"])
    predictions = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], prediction=predictions), index=test.index)
    precision = precision_score(test["target"], predictions)
    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```
